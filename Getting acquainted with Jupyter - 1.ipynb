# Learning Usage of Jupyter Notebook

## Simple Plotting Example

This is a plot of $sin$ function
$$
f(x) = \frac{\sin(x)}{x}
$$

```python
%matplotlib inline
% config InlineBackend.figure_format = 'svg'
import numpy as np
import matplotlib.pyplot as plt
pi = np.pi
x = np.linspace(-4*pi,4*pi,100)
y = np.sin(x)/x
plt.plot(x,y,label=r'y(x) = $\frac{\sin(x)}{x}$')
plt.legend(loc = 'best',fontsize = '13')
plt.title('Sin Function')
plt.axhline(0,color = 'black')
plt.grid()
plt.show()
```

# Learning Jupyter Notebook - II
## Plotting Example - II

Plot the following functions.

1. $$ y = \log(x) $$
2. $$ y = e^{-x} $$


```python
%matplotlib inline
%config InlineBackend.figure_format = 'svg'

import numpy as np
import matplotlib.pyplot as plt

x2 = np.linspace(0.1,5,100)
y2 = np.log(x2)
y3 = np.exp(-x2)
plt.grid()

plt.plot(x2,y2,label = 'Log function')
plt.plot(x2,y3,'--r',label = 'Exp function')
plt.legend(loc = 'best',fontsize = 13)
plt.show()
```
